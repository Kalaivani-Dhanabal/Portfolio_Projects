```python
# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv(r'C:\Users\13176\Downloads\movies.csv')
```

```python
df.head()
```

```python
#Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}.{}%'.format(col,pct_missing))
    
```

```python
# Dropping the missing data 
df = df.dropna(how='any',axis=0)
```

```python
df.dtypes
```

```python
#Changing Datatypes for the columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.head()
```

```python
#Create correct Year Column

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[1].astype(str).str[:4]
 df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
pd.set_option('display.max_rows', None)
df.head()
```

```python
#Drop any Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

```

```python
#Budget high Correlation
#Company HIgh Correlation
```

```python
#Scatter Plot with Budget with Gross

plt.scatter(x =df['budget'],y = df['gross'])

plt.title('Budget VS Gross Earnings')

plt.xlabel('Budget for film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] =  df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
#High Correlation between budget and gross


correlation_matrix = df_numerized.corr(method = "pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlational Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
